```python
import numpy as np
import cv2
%matplotlib inline
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
import torch.nn.functional as F
import torchvision.transforms as transforms
import torchvision.models as models

#cat 0
#dog 1
#apple 2
#cup 3


img = cv2.imread('/home/QuickDraw/image/apple28.png')
img = cv2.bitwise_not(img)
np.set_printoptions(threshold=np.inf)
#print(img)
#newimg = img.reshape(-1)
#print(newimg)
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
#print(gray)
#newimg = newimg.reshape(28,28)
plt.imshow(gray,cmap="gray")
plt.show()
```

```python
def judge(netPred):
    if netPred == [0]:
        return "cat"
    if netPred == [1]:
        return "dog"
    if netPred == [2]:
        return "apple"
    if netPred == [3]:
        return "cup"
```

```python
num_classes = 4

model_ft = models.resnet18(pretrained=False)
# fc層を置き換える
num_features = model_ft.fc.in_features
model_ft.fc = nn.Linear(num_features, num_classes)
model_ft.conv1 = nn.Conv2d(1, 64, 3, padding=1)
model_ft.avgpool = nn.AdaptiveAvgPool2d((1, 1))

gray = torch.from_numpy(gray/255)
gray = gray.view(-1,1, 28, 28)
gray = gray.float()
# モデル定義
# パラメータの読み込み
param = torch.load('quickdrawResnet.model')
device = 'cuda' if torch.cuda.is_available() else 'cpu'
#print(gray)
gray = gray.to(device)
#print(device)
#device = 'cpu'

net = model_ft.to(device)
net.load_state_dict(param)
# 評価モードにする
net = net.eval()
with torch.no_grad():
    output = net(gray)
    softmax = F.softmax(output)
    print("softmax")
    print(softmax)
    print(output)
    pred = [int(l.argmax()) for l in output]
    print(judge(pred))
```

```python
img = cv2.imread('/home/QuickDraw/image/cup.png')
img = cv2.resize(img, dsize=(28, 28))
img = cv2.bitwise_not(img)
print(img.shape)
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray,cmap="gray")
plt.show()
gray = torch.from_numpy(gray/255)
gray = gray.view(-1,1, 28, 28)
gray = gray.float()

gray = gray.to(device)
with torch.no_grad():
    output = net(gray)
    print(output)
    pred = [int(l.argmax()) for l in output]
    print(judge(pred))
```
