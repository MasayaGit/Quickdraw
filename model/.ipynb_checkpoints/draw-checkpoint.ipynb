```python
#https://github.com/uchidama/HandWriteDigitOnHtmlCanvasPredictionS
from IPython.core.display import HTML
import IPython 
HTML('''
<canvas id="canvas" height="280px" width="280px" style="border: 1px solid;"></canvas>
<p>
    <button id="clear">Clear</button>
    <button id="submit">Save image to variable</button>
</p>
<p id="msg"></p>
<script>
    var kernel = IPython.notebook.kernel;

    var config = {
        "linesize": 10,
        "linecolor": "#000000"
    }

    var mouse = {
        "X": null,
        "Y": null,
    }

    var clear = document.getElementById("clear");
    var submit = document.getElementById("submit");
    var canvas = document.getElementById("canvas");
    var ctx = canvas.getContext("2d");

    clear.addEventListener("click", function(){
        ctx.clearRect(0, 0, canvas.width, canvas.height);
    });

    submit.addEventListener("click", function(){
        var variable_value = 'base64_img';
        kernel.execute(variable_value + " = '" + canvas.toDataURL() + "'");
        msg.textContent = "Success: " + "image -> " + variable_value;
        Jupyter.notebook.execute_cells([2])
    });

    canvas.addEventListener("mouseup", drawEnd, false);
    canvas.addEventListener("mouseout", drawEnd, false);
    
    canvas.addEventListener("mousemove", function(e){
        if (e.buttons === 1 || e.witch === 1) {
            var rect = e.target.getBoundingClientRect();
            var X = e.clientX - rect.left;
            var Y = e.clientY - rect.top;
            draw(X, Y);
        };
    });
 
    canvas.addEventListener("mousedown", function(e){
        if (e.button === 0) {
            var rect = e.target.getBoundingClientRect();
            var X = e.clientX - rect.left;
            var Y = e.clientY - rect.top;
            draw(X, Y);
        }
    });

    function draw(X, Y) {
        ctx.beginPath();
        if (mouse.X === null) {
            ctx.moveTo(X, Y);
        } else {
            ctx.moveTo(mouse.X, mouse.Y);
        }
        ctx.lineTo(X, Y);
        
        ctx.lineCap = "round";
        ctx.lineWidth = config.linesize * 2;
        ctx.strokeStyle = config.linecolor;
        ctx.stroke();

        mouse.X = X;
        mouse.Y = Y;
    };
 
    function drawEnd() {
        mouse.X = null;
        mouse.Y = null;
    }
</script>
''')
```

```python
import base64
import numpy as np
from io import BytesIO
from PIL import Image
from PIL import ImageOps
import matplotlib.pyplot as plt
%matplotlib inline

base64_img = base64_img.split(",")[-1]

np.set_printoptions(threshold=10000)

img = Image.open(BytesIO(base64.b64decode(base64_img))).resize((28,28))
img = np.asarray(img)
#print(img)
plt.imshow(img)
plt.show()

def judge(netPred):
    if netPred == [0]:
        return "cat"
    if netPred == [1]:
        return "dog"
    if netPred == [2]:
        return "apple"
    if netPred == [3]:
        return "cup"
    
import numpy as np
import cv2
%matplotlib inline
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
import torch.nn.functional as F
import torchvision.transforms as transforms
import torchvision.models as models

num_classes = 4

model_ft = models.resnet18(pretrained=False)
# fc層を置き換える
num_features = model_ft.fc.in_features
model_ft.fc = nn.Linear(num_features, num_classes)
model_ft.conv1 = nn.Conv2d(1, 64, 3, padding=1)
model_ft.avgpool = nn.AdaptiveAvgPool2d((1, 1))

param = torch.load('quickdrawResnet.model')
device = 'cuda' if torch.cuda.is_available() else 'cpu'
#print(gray)
#img = cv2.imread(img)
#img = cv2.bitwise_not(img)
print(img)
#0が黒　２５５が白
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray,cmap="gray")
plt.show()

gray = torch.from_numpy(gray/255)
gray = gray.view(-1,1, 28, 28)
gray = gray.float()
gray = gray.to(device)

net = model_ft.to(device)
net.load_state_dict(param)
# 評価モードにする
net = net.eval()
with torch.no_grad():
    output = net(gray)
    print(output)
    pred = [int(l.argmax()) for l in output]
    print(judge(pred))
```
